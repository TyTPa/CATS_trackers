# Загрузка данных

```python
import pandas as pd
import gspread
from google.auth import default
from google.auth.transport.requests import Request
from google.colab import auth

# Аутентификация и получение учетных данных
auth.authenticate_user()
creds, _ = default()
gc = gspread.authorize(creds)

# Откройте Google Таблицу по её URL
spreadsheet_url = "https://docs.google.com/spreadsheets/d/1ZZIPE6JrIzLQskv37z_UXg0YW7NxmrfeyUuaDyNGL6I/edit?gid=0#gid=0/edit"
sh = gc.open_by_url(spreadsheet_url)

# Выберите первый лист таблицы
wks = sh.sheet1

# Преобразование данных в DataFrame
data = wks.get_all_records()
df = pd.DataFrame(data)


# Вывод первых 10 строк
print(df.head(10))
```

```python
# Преобразование столбцов в нужные типы данных
df['дата'] = pd.to_datetime(df['дата'], format='%d.%m.%Y', errors='coerce').dt.strftime('%d.%m.%Y')
df['время'] = pd.to_datetime(df['время'], format='%H:%M', errors='coerce').dt.strftime('%H:%M')
df['глюкоза'] = pd.to_numeric(df['глюкоза'].str.replace(',', '.'), errors='coerce')
df['доза'] = pd.to_numeric(df['доза'], errors='coerce')
df['укол время'] = pd.to_datetime(df['укол время'], format='%H:%M:%S', errors='coerce').dt.strftime('%H:%M')
df['Стул'] = df['Стул'].apply(lambda x: 1 if x == 'был' else 0)

# Добавление признака "часть дня"
df['часть дня'] = df['время'].apply(lambda x: 'утро' if pd.to_datetime(x, format='%H:%M').time() < pd.to_datetime('14:00').time() else 'вечер')

# Вывод первых 10 строк
print(df.head(10))
```

Проверка формата данных

```python
# Преобразование столбцов в нужные типы данных
df['дата'] = pd.to_datetime(df['дата'], format='%d.%m.%Y', errors='coerce').dt.strftime('%d.%m.%Y')
df['время'] = pd.to_datetime(df['время'], format='%H:%M', errors='coerce').dt.strftime('%H:%M')
df['глюкоза'] = df['глюкоза'].astype(str).str.replace(',', '.').astype(float)
df['доза'] = pd.to_numeric(df['доза'], errors='coerce')
df['укол время'] = pd.to_datetime(df['укол время'], format='%H:%M', errors='coerce').dt.strftime('%H:%M')
df['Стул'] = df['Стул'].apply(lambda x: 1 if x == 'был' else 0)
df.head()
```

```python
# Преобразование типов данных
df['глюкоза'] = pd.to_numeric(df['глюкоза'])
df['доза'] = pd.to_numeric(df['доза'])
df['дата'] = pd.to_datetime(df['дата'], format='%d.%m.%Y')

# Разделение на утро и вечер
df_утро = df[df['время'] == 'утро']
df_вечер = df[df['время'] == 'вечер']

# Группировка данных по дате
df_утро_grouped = df_утро.groupby('дата').agg({'глюкоза': 'mean', 'доза': 'mean', 'Стул': lambda x: x.mode()[0]})
df_вечер_grouped = df_вечер.groupby('дата').agg({'глюкоза': 'mean', 'доза': 'mean', 'Стул': lambda x: x.mode()[0]})

# Создание графиков
fig, axs = plt.subplots(2, 1, figsize=(10, 10))

# Утро
axs[0].plot(df_утро_grouped.index, df_утро_grouped['глюкоза'], label='Глюкоза (утро)', color='blue', marker='o')
axs[0].plot(df_утро_grouped.index, df_утро_grouped['доза'], label='Доза (утро)', color='orange', marker='o')
axs[0].set_title('Утро: Глюкоза и Доза')
axs[0].set_xlabel('Дата')
axs[0].set_ylabel('Значение')
axs[0].legend()
axs[0].grid()
axs[0].set_xticks(df_утро_grouped.index)
axs[0].set_xticklabels(df_утро_grouped.index.strftime('%d.%m'), rotation=45)

# Вечер
axs[1].plot(df_вечер_grouped.index, df_вечер_grouped['глюкоза'], label='Глюкоза (вечер)', color='blue', marker='o')
axs[1].plot(df_вечер_grouped.index, df_вечер_grouped['доза'], label='Доза (вечер)', color='orange', marker='o')
axs[1].set_title('Вечер: Глюкоза и Доза')
axs[1].set_xlabel('Дата')
axs[1].set_ylabel('Значение')
axs[1].legend()
axs[1].grid()
axs[1].set_xticks(df_вечер_grouped.index)
axs[1].set_xticklabels(df_вечер_grouped.index.strftime('%d.%m'), rotation=45)

plt.tight_layout()
plt.show()
```
